# A.01 Empirirical Rule

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Data for standard normal curve
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# Colors (consistent between legend and plot)
c68 = "#66c2a5"   # green
c95 = "#fc8d62"   # orange
c997 = "#8da0cb"  # blue

plt.figure(figsize=(10, 6))

# Plot the density curve
plt.plot(x, y, color="black", linewidth=2)

# Draw from OUTSIDE -> INSIDE with solid colors (alpha=1) to avoid color mixing
# 99.7% region: [-3, -2] U [2, 3]
plt.fill_between(x, y, where=((x >= -3) & (x <= -2)) | ((x >= 2) & (x <= 3)), color=c997, label="99.7% within 3\u03C3")

# 95% region (excluding inner 68%): [-2, -1] U [1, 2]
plt.fill_between(x, y, where=((x >= -2) & (x <= -1)) | ((x >= 1) & (x <= 2)), color=c95, label="95% within 2\u03C3")

# 68% region: [-1, 1]
plt.fill_between(x, y, where=(x >= -1) & (x <= 1), color=c68, label="68% within 1\u03C3")

# Vertical line at mean
plt.axvline(0, color="black", linestyle="--", linewidth=1)

# Labels inside regions
# 68% label (center)
plt.text(0, norm.pdf(0, 0, 1)*0.6, "68%", color="white", ha="center", va="center", fontsize=12, fontweight="bold")

# 95% labels (left/right)
plt.text(-1.5, norm.pdf(-1.5, 0, 1)*0.5, "95%", color="white", ha="center", va="center", fontsize=11, fontweight="bold", rotation=0)
plt.text( 1.5, norm.pdf( 1.5, 0, 1)*0.5, "95%", color="white", ha="center", va="center", fontsize=11, fontweight="bold", rotation=0)

# 99.7% labels (left/right outer tails)
plt.text(-2.3, norm.pdf(-2.5, 0, 1)*0.45, "99.7%", color="white", ha="center", va="center", fontsize=10, fontweight="bold")
plt.text( 2.3, norm.pdf( 2.5, 0, 1)*0.45, "99.7%", color="white", ha="center", va="center", fontsize=10, fontweight="bold")

# Titles and axes
plt.title("Empirical Rule (68–95–99.7 Rule)", fontsize=14)
plt.xlabel("Standard Deviations from Mean (σ)", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(loc="upper left")
plt.tight_layout()

# Save figure
plt.savefig('chp1_empirical_rule.png', dpi=300)
plt.show()
```
